##  Analysis through Visualization

**Univariate Distribution**

Univariate Distribution is used to analyse the distribution for single variable 

```
sns.distplot(ALT['city_mpg'])
plt.show()
```

This distplot can only plotted for numerical/contineous variables not able to plot for categorical variables

```

sns.distplot(ALT['normalized_losses'],kde=0,rug=True)
plt.show()
```

In the above code snippet. We can see lines at the bottom of each bins which indicates the frequency of the variable , we can relate this as bin height increases the concentration of the lines in that bin also increases 

**Bivariate Distribution**

It is used to analyse any Two contineous variables 

```
sns.jointplot(ALT["engine_size"],ALT['horsepower'])
plt.show()
```

```
sns.jointplot(ALT['engine_size'],ALT['horsepower'],kind='hex')
```

We used Two kind of plots here ,Basically bivariate plots are used to see the correlation between Two variables ,By looking above plots we can conclude that both variables have a posivite correlation.
By above plots we can get insights for three things
     
     1) Distribution of Enigine_size
     2) Distribution of Horse power
     3) Correlation betweeen this two features
In second plot I used hexagon instead of the scatter as the concentration of hexagon is more we can conclude observation at such points are more     

 **Multivariate distribution**

This kind of plots are used to analyse more than 2 features at a time in single plot

```
sns.pairplot(ALT[['price','normalized_losses','horsepower','engine_size']])
```

In above plot,we clearly see that the diagonal plots are histograms which means we can analyse  all the features distribution in only one plots and also we can get the correlation of features in single plot itself.if we observe carefully the Upper and Lower triangles are nothing but mirror image  of each others



Till now we only focused on numerical variable.Now lets see about categoriacal variables

```
sns.swarmplot(ALT['fuel_type'],ALT['horsepower'])
plt.show()
```

By above plot we can see layer wise distribution of data the insights that can be drawn from the plots are

1) We have an imbalance dataset

2) More vehicles have their horsepower range between ~ (75-125) 

```
sns.boxplot(ALT['number_of_doors'],ALT['horsepower'])
```

The above plot is known as boxplots used for statistic purpose 

1) The top and bottom line of the viscous tells us the maximum and minimum value
from above plot we see for 2 door car the maximum value is around 220 and mimimum value is exactly 50, and for 4 door car the maximum value is around 170 and minimum value is around 50

2) The 25th percentile,75th percentile and the median/50th percentile can be seen

3) The points outside the plots are the outliers.

```
sns.factorplot(x='body_style',y='horsepower',data=ALT,col='fuel_type',hue='aspiration',kind='violin')
sns.factorplot(x='body_style',y='horsepower',data=ALT,col='fuel_type',hue='aspiration',kind='bar')
sns.factorplot(x='body_style',y='horsepower',data=ALT,hue='aspiration',col='fuel_type',kind='point')
plt.show()
```

Factorplots are used to draw multi panel categorical plots using array of values
we get more insights by such kind of the plots its a advance level of plotting



```
s=ALT.groupby('body_style',axis=0).sum()
```

```
s.head()
```

```

explode=(0,0.1,0,0,0)
s['horsepower'].plot(kind='pie',
                     shadow=True,
                     explode=explode,
                     autopct='%1.1f%%'
                     )
plt.title("price of each body style")
plt.show()
```

We are able to see that horsepower of sedan is more.
