# visualization using seaborn 

imported all required library for this work

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
import pandas as pd
```

imported a dataset of automobile vehicle

```python
a=pd.read_csv('auto.csv')
```

```python
a.head(5)
```

The glimpse of dataset

```python
sns.lineplot(x='length',y="price",data=a,hue='fuel-type',style='fuel-type',markers=True)
plt.show()
```

```python
sns.barplot(x='body-style',y='price',data=a)
plt.show()
```

```python
sns.barplot(x='fuel-type',y='price',data=a)
plt.show()
```

```python
sns.barplot(x='fuel-type',y='length',data=a)
plt.show()
```

```python
sns.barplot(x='engine-size',y='price',data=a)
plt.show()
```

```python
sns.barplot(x='num-of-cylinders',y='price',data=a)
plt.show()
```

```python
sns.barplot(x='aspiration',y='price',data=a)
plt.show()
```

```python
sns.barplot(x='num-of-doors',y='price',data=a)
plt.show()
```

```python
sns.barplot(x='drive-wheels',y='price',data=a,hue="aspiration")
```

```python
sns.barplot(x='fuel-type',y='price',hue='engine-size',data=a)
plt.show()
```

```python
sns.distplot(a['price'],color='red',bins=10)
plt.show()
```

```python
sns.distplot(a['length'],vertical=True,color='green')
```

```python
sns.jointplot(x='price',y='length',data=a,kind='reg')
plt.show()
```

```python
sns.boxplot(x="fuel-type",y='price',data=a,hue='engine-size')
plt.show()
```

```python
sns.boxplot(x='num-of-cylinders',y='price',data=a,palette='Set1',linewidth='3')
plt.show()
```
